```python
nifty = yf.Ticker("^NSEI")
```

```python
nifty = nifty.history(period = "max")
```

```python
del nifty["Dividends"]
del nifty["Stock Splits"]
```

```python
nifty["Tomorrow"] = nifty["Close"].shift(-1)
```

```python
nifty
```

```python
nifty["Target"] = (nifty["Tomorrow"] > nifty["Close"]).astype(int)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = nifty.iloc[:-100]
test = nifty.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index = test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis = 1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds, index = test.index, name = "Predictions")
  combined = pd.concat([test["Target"], preds], axis = 1)
  return combined
```

```python
def backtest(data, model, predictors, start = 1250, step = 250):
  all_predictions = []

  for i in range(start, data.shape[0], step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = predict(train, test, predictors, model)
    all_predictions.append(predictions)
  return pd.concat(all_predictions)

```

```python
predictions = backtest(nifty, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
horizons = [2, 5, 60, 250, 1000]

new_predictors = []

for horizon in horizons:
  rolling_avg = nifty.rolling(horizon).mean()

  ratio_col = f"Close_Ratio_{horizon}"
  nifty[ratio_col] = nifty["Close"] / rolling_avg["Close"]

  trend_col = f"Trend_{horizon}"
  nifty[trend_col] = nifty.shift(1).rolling(horizon).sum()["Target"]

  new_predictors += [ratio_col, trend_col]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))

plt.plot(nifty["Close"], label="NIFTY Closing Price", alpha=0.8)

for horizon in [60, 250, 1000]:
    rolling_avg = nifty["Close"].rolling(window=horizon).mean()
    plt.plot(rolling_avg, label=f"{horizon}-Day Moving Average", linestyle="--")

plt.title("NIFTY Closing Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict_proba(test[predictors])[:,1]
  preds[preds >= 0.6] = 1
  preds[preds < 0.6] = 0
  preds = pd.Series(preds, index = test.index, name = "Predictions")
  combined = pd.concat([test["Target"], preds], axis = 1)
  return combined
```
